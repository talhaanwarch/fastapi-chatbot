```python
import os
import requests
import pandas as pd
import re
from tqdm.notebook import tqdm

# Create directory if it doesn't exist
os.makedirs('data', exist_ok=True)

# Create list to store file paths
file_paths = []

for index, row in tqdm(df.iterrows(), total=df.shape[0]):
    try:
        # Clean title to create safe filename
        clean_title = re.sub(r'[^\w\s-]', '', row['Title']).strip()
        clean_title = re.sub(r'[-\s]+', '_', clean_title)
        
        # Create filename
        filename = f"{row['Year']}_{clean_title[:50]}.pdf"  # Limit title length
        filepath = os.path.join('data', filename)
        
        # Check if file already exists
        if os.path.exists(filepath):
            file_paths.append(filepath)
            continue  # Skip download
        
        # Download PDF
        response = requests.get(row['PDF URL'], timeout=10)
        response.raise_for_status()  # Raise exception for HTTP errors
        
        # Save file
        with open(filepath, 'wb') as f:
            f.write(response.content)
        
        file_paths.append(filepath)
        
    except Exception as e:
        print(f"Error downloading {row['Title']} {row['PDF URL']}: {e}")
        file_paths.append(None)

# Add file paths to DataFrame
df['Local_File_Path'] = file_paths

# Save updated DataFrame (optional)
df.to_csv('document_index.csv', index=False)
```

```python
import pandas as pd
from pypdf import PdfReader
from pathlib import Path

def count_pdf_pages(pdf_path):
    """
    Counts the number of pages in a PDF file.
    
    Args:
        pdf_path (str): Path to the PDF file
        
    Returns:
        int: Number of pages, or 0 if file can't be read
    """
    try:
        # Verify file exists
        if not Path(pdf_path).exists():
            print(f"File not found: {pdf_path}")
            return 0
        
        with open(pdf_path, 'rb') as f:
            pdf = PdfReader(f)
            return len(pdf.pages)
            
    except Exception as e:
        print(f"Error processing {pdf_path}: {str(e)}")
        return 0


# Apply the function to create new column
df['page_count'] = df['Local_File_Path'].apply(count_pdf_pages)
```

```python
import pandas as pd
from concurrent.futures import ThreadPoolExecutor, as_completed

# Function to extract text from PDF and create new rows
def extract_text_and_create_rows(row):
    local_file_path = row['Local_File_Path']
    title = row['Title']
    year = row['Year']
    pdf_url = row['PDF URL']
    
    # Open the PDF file
    with open(local_file_path, 'rb') as file:
        reader = PdfReader(file)
        num_pages = len(reader.pages)
        
        # Extract text from each page
        rows = []
        for page_num in range(num_pages):
            page = reader.pages[page_num]
            text = page.extract_text()
            
            # Format the content
            #formatted_text = f"# Title: {title}\n## Year: ({year})\n## Content:\n{text}\n## Source:\n### Page number {page_num + 1} \n### URL: [PDF]({pdf_url})"
            
            # Create a new row
            new_row = {
                'Year': year,
                'Title': title,
                'Sales No': row['Sales No'],
                'PDF URL': pdf_url,
                'Local_File_Path': local_file_path,
                'Page Number': page_num + 1,
                'Content': text
            }
            
            rows.append(new_row)
        
        return rows

# Function to process rows concurrently while preserving order
def process_concurrently(df):
    results = []
    with ThreadPoolExecutor(max_workers=24) as executor:
        # Submit tasks to the executor
        future_to_index = {executor.submit(extract_text_and_create_rows, row): idx for idx, row in df.iterrows()}
        
        # Collect results in the order of the original DataFrame
        for future in as_completed(future_to_index):
            idx = future_to_index[future]
            try:
                rows = future.result()
                results.extend(rows)
            except Exception as e:
                print(f"Error processing row {idx}: {e}")
    
    # Sort results by the original index to maintain order
    results.sort(key=lambda x: df.index[df['Local_File_Path'] == x['Local_File_Path']].tolist()[0])
    return results

# Process the DataFrame concurrently
result_rows = process_concurrently(df)

# Create a new DataFrame to store the results
result_df = pd.DataFrame(result_rows)
len(result_df)
```

```python
import tiktoken
def num_tokens_from_string(string: str, encoding_name='cl100k_base') -> int:
    """Returns the number of tokens in a text string."""
    encoding = tiktoken.get_encoding(encoding_name)
    num_tokens = len(encoding.encode(string))
    return num_tokens
```

```python
from langchain_core.documents import Document
# Function to modify a single Document
def modify_document(doc):
    new_page_content = (
        f"# Title: {doc.metadata['title']}\n"
        f"## Year: ({doc.metadata['year']})\n"
        f"## Content:\n"
        f"{doc.page_content}\n"
        f"## Page URL: {doc.metadata['pdf_url']} \n"
        f"## Page Number: {doc.metadata['page_number']}"
    )
    return Document(
        metadata=doc.metadata,
        page_content=new_page_content
    )

# Apply the modification to all items in the list
docs = [modify_document(doc) for doc in docs]
```
